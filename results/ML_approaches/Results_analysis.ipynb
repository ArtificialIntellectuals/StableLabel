```python
df_list = []

for i in [31,32,33,34,35,36]:
  with open(f"{path}external_{i}.pkl", 'rb') as f:
      run = pd.DataFrame(pickle.load(f))
      run["0_count"].apply(lambda x: x-101)
      df_list.append(run)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings

# Your code here

with warnings.catch_warnings():
  warnings.simplefilter("ignore")

  for id in df_list[0].id.unique():
    df_stats=df_list[0][df_list[0].id == id][['p','r','f1']]
    for df in df_list[1:]:
      df_stats=df_stats.append(df[df.id == id][['p','r','f1']])
    print(id)
    print(df_stats.describe())
    bplt =sns.boxplot(df_stats)

    bplt.plot()
    plt.title(f'{id}')
    plt.show()
```

```python
import warnings

# Your code here

with warnings.catch_warnings():
  warnings.simplefilter("ignore")
  df_stats_list=[]

  for id in df_list[0].id.unique():
    if id!="Embeddings: MLP (ReLu + SGD) and hidden_sizes 10-10":
      df_stats=df_list[0][df_list[0].id == id][['p','r','f1']]
      for df in df_list[1:]:
        df_stats=df_stats.append(df[df.id == id][['id','p','r','f1']])
      df_stats_list.append(df_stats)

  for i in range(3):
    df_stats= df_stats_list[i*4]
    if i==2:
      for df in df_stats_list[i*4+1:]:
        df_stats=df_stats.append(df)
    else:
      for df in df_stats_list[i*4+1:(i+1)*4]:
        df_stats=df_stats.append(df)
    sns.set(style="whitegrid")

    # Create boxplots with nested grouping by 'id' and 'Metric'
    plt.figure(figsize=(12, 4))
    ax = sns.boxplot(x='id', y='value', hue='variable', data=pd.melt(df_stats
    [['id', 'p', 'r', 'f1']], id_vars='id'))
    ax.set_xticklabels(ax.get_xticklabels(), rotation=8)  # adjust the rotation and ha as needed

    plt.xlabel('ID')

    plt.tight_layout()
    plt.savefig(f'{path}80_nothing_something_{i}_internal.png')
    plt.show()
```
